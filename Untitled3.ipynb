```python
from datetime import datetime
from binance.client import Client
import pandas as pd
import statistics 
import sklearn
import numpy as np
import matplotlib.pyplot as plt
from joblib import dump,load
from sklearn.ensemble import RandomForestRegressor
import grindfunc



           
variables_definition=[
    {"name":"minutes","n_samples":1,"n_values":300},
    {"name":"15minutes","n_samples":15,"n_values":20},
    {"name":"hours","n_samples":60,"n_values":5},
    {"name":"6hours","n_samples":360,"n_values":5},
    ]

target_definition={"samples_from":30, "samples_to":120, "n_samples":10}

n_samples=0
n_values=0
for v in variables_definition:
    v["offset_values"]=n_values
    v["offset_samples"]=n_samples
    n_samples=n_samples+v["n_samples"]*v["n_values"]
    n_values=n_values+v["n_values"]
    
values=list(range(n_values))
df=grindfunc.import_samples("2021-03-01")

h_samples=df.values.tolist()
print("Importati ",len(h_samples),"valori da",h_samples[-1],"a",h_samples[0])
print("Variabili",n_values)
print("Giorni",grindfunc.twodec(n_samples/60/24))
print("Campioni",n_samples)
print("Punti",len(h_samples)-n_samples)
```

```python
rf = load('rf2.joblib') 
```

```python
samples=[]
for i in range(n_samples):
    #print("preshift",h_samples[-1])
    samples.insert(0,h_samples[-1][1])
    h_samples.pop(-1)

capital=100000
invested=0
capit=[]
valu=[]
predu=[]
old_capital=capital
BTC=0

count=0
while(len(h_samples)>0):
    grindfunc.calculate_values(samples, variables_definition,values)
    X=np.array([values])
    prediction = rf.predict(X)
    count+=1
    if not count%60:
        print(".",h_samples[-1][0],round(h_samples[-1][1]))
        
    capit.append(old_capital)
    predu.append(prediction[0])
    valu.append(h_samples[-1][1])
    if prediction[0]>1 and invested==0 and capital>0:
        BTC=capital/h_samples[-1][1]*(1-0.0075)
        print(h_samples[-1][0],"\tBUY\t",int(h_samples[-1][1]),"\tCapital\t",int(capital),"\tBTC\t",BTC)
        old_capital=capital
        capital=0
    if BTC>0 and (prediction[0]<0 and BTC*h_samples[-1][1]/old_capital>1.005):
        
        capital=BTC*h_samples[-1][1]
        BTC=0

        gain=(capital-old_capital)/old_capital
        print(h_samples[-1][0],"\tSELL\t",int(h_samples[-1][1]),"\tCapital\t",int(capital),"\t%\t",gain)

    samples.pop(-1)
    samples.insert(0,h_samples[-1][1]) 
    h_samples.pop(-1)
 
```

```python
print(int(capital)) 
predu2=[]
a=0
b=90000
for i in range(len(predu)):
    if(predu[i]>1.3):
        predu2.append(valu[i]+predu[i]*2000)
    elif predu[i]<0.5:
        predu2.append(valu[i]+(-0.5+predu[i])*2000)
    else :
        predu2.append(valu[i])
    
plt.figure(figsize=(20, 12))
plt.subplot(2, 1, 1)
plt.plot(capit[a:b])       
plt.subplot(2, 1, 2)
plt.plot(predu2[a:b])  
plt.subplot(2, 1, 2)
plt.plot(valu[a:b])
     
```
