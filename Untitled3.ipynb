```python
# from datetime import datetime
from binance.client import Client
import pandas as pd
import statistics 
import sklearn
import numpy as np
import matplotlib.pyplot as plt
from joblib import dump,load
from sklearn.ensemble import RandomForestRegressor
import grindfunc
           
variables_definition=[
    {"name":"minutes","n_samples":1,"n_values":60},
    {"name":"15minutes","n_samples":15,"n_values":28},
    {"name":"hours","n_samples":60,"n_values":0},
    {"name":"6hours","n_samples":360,"n_values":0},
    ]

target_definition={"samples_from":5, "samples_to":120, "n_samples":10}

n_samples=0
n_values=0
for v in variables_definition:
    v["offset_values"]=n_values
    v["offset_samples"]=n_samples
    n_samples=n_samples+v["n_samples"]*v["n_values"]
    n_values=n_values+v["n_values"]
    
values=list(range(n_values))
df=grindfunc.import_samples("Binance_BTCUSDT_minute (15).csv","2021-02-01")

h_samples=df.values.tolist()
print("Importati ",len(h_samples),"valori da",h_samples[-1],"a",h_samples[0])
print("Variabili",n_values)
print("Giorni",grindfunc.twodec(n_samples/60/24))
print("Campioni",n_samples)
print("Punti",len(h_samples)-n_samples)
```

```python
rf = load('rf4.joblib') 
```

```python
samples=[]
for i in range(n_samples):
    #print("preshift",h_samples[-1])
    samples.insert(0,h_samples[-1][1])
    h_samples.pop(-1)

capital=100000
invested=0
capit=[]
valu=[]
predu=[]
guadu=[]
BTC=0
buybu=[]
sellbu=[]

count=0
while(len(h_samples)>target_definition["samples_to"]):
    grindfunc.calculate_values(samples, variables_definition,values)
    X=np.array([values])
    prediction = rf.predict(X)
    guadagno=grindfunc.get_target(h_samples, target_definition)
    if not count%10:
        print(h_samples[-1][0],"\t",round(h_samples[-1][1]),"\t",prediction[0],"\t",guadagno)

        
    predu.append(prediction[0])
    valu.append(h_samples[-1][1])
    guadu.append(guadagno)

        
    samples.pop(-1)
    samples.insert(0,h_samples[-1][1]) 
    h_samples.pop(-1)
    count+=1

 
```

```python
df=grindfunc.import_samples("Binance_BTCUSDT_minute (14).csv","2021-02-01")

h_samples=df.values.tolist()
print(len(h_samples))

samples=[]
for i in range(n_samples):
    #print("preshift",h_samples[-1])
    samples.insert(0,h_samples[-1][1])
    h_samples.pop(-1)

print(len(h_samples))
capital=100000
invested=0
capit=[]
old_capital=capital
BTC=0
buybu=[]
buybuinst=[]
sellbu=[]
sellbuinst=[]
sellbureason=[]
vecchiu=[]
guadoz=[]
confirm=0
count=0
decrease=0
increase=0
prev_value=0
prev_gain=0
reason=""
vecchio=0
post_gain=-1
maxgain=0.5
print(len(h_samples),target_definition["samples_to"],len(predu))
while(len(h_samples)>target_definition["samples_to"]):
    X=np.array([values])
    prediction = [predu[count]]
    #if not count%10:
        #print(h_samples[-1][0],"\t",round(h_samples[-1][1]),"\t",prediction[0],"\t",round(old_capital),"\t",round(BTC*h_samples[-1][1]))
        #z=input("caz")
    
    capit.append(old_capital)
    if capital>0 and ((prediction[0]>1.6 and decrease<1 and post_gain<0) or prediction[0]>2.5):
        confirm+=1
        #print(prediction[0],confirm)
        if confirm>1:
            BTC=capital/h_samples[-1][1]*(1-0.0075)
            print(h_samples[-1][0],"\tBUY    P",prediction[0],"\t",int(h_samples[-1][1]),"\tCapital\t",int(capital))
            old_capital=capital
            capital=0
            buybu.append(prediction[0])
            buybuinst.append(count)
            confirm=0
    else:
        confirm=0
    
    
    flag_sell=False
    
    if h_samples[-1][1]<prev_value*0.999:
        decrease+=1
        #print("calina",decrease)
    else:
        decrease-=1
        if decrease<0:
            decrease=0
    
    if h_samples[-1][1]>prev_value*1.001:
        increase+=1
        #print("calina",decrease)
    else:
        increase=0
        if increase<0:
            increase=0
        
    gain=BTC*h_samples[-1][1]/old_capital
    
    if  BTC>0:
        reason=""
        if gain>maxgain:
            maxgain=gain
        
        if  (prediction[0]< 0.5 and gain>1.01 and decrease>2 ):
            flag_sell=True
            reason=reason+"PREDICT "+str(grindfunc.twodec(gain))+str(decrease)
            
        if  (gain>1.05 and decrease>3):
            flag_sell=True
            reason=reason+"5gain   "+str(grindfunc.twodec(gain))

        if gain<0.95:
            flag_sell=True
            reason=reason+"LOSS "+str(grindfunc.twodec(gain))

        if vecchio>240 and gain<1:
            flag_sell=True
            reason=reason+"VECCHIO "+str(grindfunc.twodec(gain))

        if vecchio>120 and gain>1.05 and(gain-1)<(maxgain-1)*0.4:
            flag_sell=True
            reason=reason+"DEGUADO "+str(grindfunc.twodec(gain))+" "+str(grindfunc.twodec(maxgain))

            
        if decrease>3 and gain>1.0075:
            flag_sell=True
            reason=reason+"CALA "+str(grindfunc.twodec(gain))
        
        if decrease>20 and gain>1.02:
            flag_sell=True
            reason=reason+"CALETTA "+str(grindfunc.twodec(gain))

        if decrease>7 and gain>0.995:
            flag_sell=True
            reason=reason+"CALONA "+str(grindfunc.twodec(gain))

            
        if flag_sell:
            post_gain=gain*1000
            capital=BTC*h_samples[-1][1]
            BTC=0
            decrease=0
            gain2=grindfunc.twodec(100*(capital-old_capital)/old_capital)
            print(h_samples[-1][0],"\t",reason,"\t",int(h_samples[-1][1]),"\tCapital\t",int(capital),"\t\t\t%\t",gain2)
            sellbu.append(prediction[0])
            sellbuinst.append(count)
            sellbureason.append(reason)
            vecchiu.append(vecchio)
            reason=""

    
    if BTC:
        vecchio=vecchio+1
    else:
        vecchio=0
    
    guadoz.append(str(round(gain,3))+" "+str(decrease)+" "+str(increase))
    prev_gain=gain
    prev_value=h_samples[-1][1]
    samples.pop(-1)
    samples.insert(0,h_samples[-1][1]) 
    h_samples.pop(-1)
    count+=1
    post_gain=post_gain-1
```
